```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_white = np.ones((400,450,3), np.uint8) * 255
blank_black = np.zeros((400,450,3), np.uint8) * 255

# hist = cv2.calcHist([blank_white], [0], None, [255], [0,256])
hist = cv2.calcHist([blank_black], [0], None, [255], [0,256])
plt.plot(hist)
plt.show


cv2.imshow("blank white", blank_white)
cv2.imshow("blank black", blank_black)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))

img1 = cv2.imread("Cat 1.jpg")
img2 = cv2.imread("Cat 2.jpg")
img3 = cv2.imread("Cat 3.jpg")


hist1 = cv2.calcHist([img1], [0], None, [255], [0,256])
hist2 = cv2.calcHist([img2], [0], None, [255], [0,256])
hist3 = cv2.calcHist([img3], [0], None, [255], [0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)
plt.show


cv2.imshow("Cat 1", img1)
cv2.imshow("Cat 2", img2)
cv2.imshow("Cat 3", img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("zeeshan_fareed.jpg")
img1 = cv2.resize(img1, (400,450))

b,g,r = cv2.split(img1)

hist1 = cv2.calcHist([img1], [0], None, [255], [0,256])
hist2 = cv2.calcHist([img2], [0], None, [255], [0,256])
hist3 = cv2.calcHist([img3], [0], None, [255], [0,256])

plt.subplot(3,1,1)
plt.plot(hist1)

plt.subplot(3,1,2)
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)
plt.show


cv2.imshow("Image 1", img1)
cv2.imshow("Image B", b)
cv2.imshow("Image G", g)
cv2.imshow("Image R", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## combine Histogram

```python
img1 = cv2.imread("zeeshan_fareed.jpg")
img1 = cv2.resize(img1, (400,450))

b,g,r = cv2.split(img1)

plt.hist(b.ravel(), 255, (0,255), color = "blue")
plt.hist(g.ravel(), 255, (0,255), color = "green")
plt.hist(r.ravel(), 255, (0,255), color = "red")

plt.show


cv2.imshow("Image 1", img1)
cv2.imshow("Image B", b)
cv2.imshow("Image G", g)
cv2.imshow("Image R", r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
```

## Adaptive Equalizer

```python
img = cv2.imread("hist.jpg")
img = cv2.resize(img, (500,450))

gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry], [0], None, [255], [0,255])

eq = cv2.equalizeHist(gry)
hist1 = cv2.calcHist([eq], [0], None, [255], [0,255])


plt.plot(hist1, c = "red")
plt.plot(hist)
plt.show()

cv2.imshow("Image", img)
cv2.imshow("gray", eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## clahe equalizer

```python
img = cv2.imread("clahe.jpg")
img = cv2.resize(img, (500,450))

gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gry], [0], None, [255], [0,255])

eq = cv2.equalizeHist(gry)
hist1 = cv2.calcHist([eq], [0], None, [255], [0,255])

cl = cv2.createCLAHE(clipLimit = 1 , tileGridSize = (8,8))
f = cl.apply(gry)

plt.plot(hist1, c = "red")
plt.plot(f, c = "green")
plt.plot(hist)
plt.show()

cv2.imshow("Image", img)
cv2.imshow("Equalizer", eq)
cv2.imshow("Final Image", f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
